```python
# assignment: how to change x and y axis titles?

import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4, 5], [10, 20, 30, 40, 50])
plt.title("servived")
plt.show()
```

```python
g= sns.FacetGrid(titanic, hue="survived", col="pclass", margin_titles=True,
              palette={1:"yellowgreen", 0:"gray"})
g= g.map(plt.scatter, "fare", "age", edgecolor="y").add_legend();
```

```python
g= sns.FacetGrid(titanic, hue="survived", col="sex", margin_titles=True,
              palette={1:"yellowgreen", 0:"gray"})
g= g.map(plt.scatter, "fare", "age", edgecolor="y").add_legend();
```

```python
g= sns.FacetGrid(titanic, hue="survived", col="who", margin_titles=True,
              palette={1:"yellowgreen", 0:"gray"})
g= g.map(plt.scatter, "fare", "age", edgecolor="b").add_legend();
```
